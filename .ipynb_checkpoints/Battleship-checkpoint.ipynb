```python
import numpy

board_size = 10

ship_types = {
        "carrier": 5,
        "battleship": 4,
        "cruiser": 3,
        "submarine": 3,
        "destroyer": 2
        }

directions = ["up", "down", "left", "right"]

def init_board(size):
    return numpy.zeros((size, size))

def init_user():
    return { "ships": {}, "board": init_board(board_size), "guess_board": init_board(board_size) }

def parse_location(location):
    return [int(location[1:]), ord(location[0].lower()) - 97]

def validate_location(location):
    letter = location[0]
    number = location[1:]
    if letter.isalpha() and number.isdigit():
        letter.lower()
        number = int(number)
        return (letter.isalpha() and ord(letter) > 96 and ord(letter) < 97 + board_size and number > -1 and number < board_size)
    return False

#def validate_placement(location, direction, board):

def place_ships(user):
    print ("give a letter (a-" + chr(board_size + 96) + ") and number (0-" + str(board_size - 1) + ") to indicate placement location")
    print_board(user["board"])
    for ship in ship_types:
        valid = False
        while (not valid):
            location = input("where would you like to place your " + ship + "? (" + str(ship_types[ship]) + " spaces)")
            if (not validate_location(location)):
                print("invalid location")
                continue
            direction = input("what direction would you like it to face (up, down, left, right)")
            if (direction not in directions):
                print ("invalid direction")
                continue
            index = parse_location(location)
            state = index[:]
            for i in range(ship_types[ship]):
                if (state[0] < board_size and state[1] < board_size and
                        state[0] > -1 and state[0] > -1 and
                        user["board"][state[0], state[1]] == 0
                        ):
                    valid = True
                    user["board"][state[0], state[1]] = 1
                    if direction == "up":
                        state[0] -= 1
                    elif direction == "down":
                        state[0] += 1
                    elif direction == "left":
                        state[1] -= 1
                    elif direction == "right":
                        state[1] += 1
                else:
                    print ("invalid location")
                # make the user redo the location
            user["ships"][ship] = (index, direction)
            print_board(user["board"])

def print_board(board):
    temp  = " "
    for i in range(board_size):
        temp += str(chr(65 + i)) + " "
    print(temp)
    for i in range(board_size):
        temp = ""
        for j in range(board_size):
            if j == 0:
                temp += str(i)
            if board[i][j] == 0:
                temp += ("‚¨ú")
            elif board[i][j] == 1:
                temp += ("‚¨õ")
            elif board[i][j] == 2:
                temp += ("‚ùå")
            elif board[i][j] == 3:
                temp += ("üö´")
        print(temp)

def turn(user, enemy):
    print_board(user["board"])
    print_board(user["guess_board"])
    location = input("where would you like to hit?")
    index = parse_location(location)
    if (enemy["board"][index[0], index[1]] == 1):
        print("HIT!!!!!")
        enemy["board"][index[0], index[1]] = 2
        user["guess_board"][index[0], index[1]] = 2
    else:
        print("MISS")
        user["guess_board"][index[0], index[1]] = 3

def init_game():
    state = "playing"
    while (state == "playing"):
        print("PLAYER ONE TURN")
        turn(player_one, player_two)
        print("---------------------------------------------")
        print("player two turn")
        turn(player_two, player_one)

player_one = init_user()
player_two = init_user()

print("place ships for player 1...")
place_ships(player_one)

print("place ships for player 2...")
place_ships(player_two)

init_game()
```
