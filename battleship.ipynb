```python
#import numpy as np

class Ship:
    HORIZONTAL = 0
    VERTICAL = 1

    def __init__(self, leng):
        self.leng = leng
        self.dire = -1
        self.row = -1
        self.col = -1

    def getLength(self) :
        return self.leng
    def getRow(self) :
        return self.row
    def getCol(self) :
        return self.col
    def getDir(self) :
        return self.dire

    def setRow(self, x) :
        self.row = x
    def setCol(self, y) :
        self.col = y
    def setDir(self, z) :
        self.dire = z

    def isDirSet(self) :
        if(self.dire == -1) :
            return False
        return True
    def isLocSet(self) :
        if(self.row == -1 or self.col == -1) :
            return False
        return True

    def printDir(self) :
        if(self.dire == 0) :
            return "horizontal"
        elif(self.dire == 1) :
            return "vertical"
        else :
            return "unset"
    def printLoc(self) :
        if(self.isLocSet()) :
            return "(" + str(self.getRow()) + ", " + str(self.getCol()) + ")"
        return "unset location"
    def printShip(self) :
        return self.printDir() + " ship of length " + str(self.getLength()) + " at " + self.printLoc()

class Location :
    UNGUESSED = 0
    HIT = 1
    MISSED = 2

    def __init__(self) :
        self.exist = False
        self.status = 0

    def isHit(self) :
        if(self.status == 1) :
            return True
        return False
    def isMissed(self) :
        if(self.status == 2) :
            return True
        return False
    def isUnguessed(self) :
        if(self.status == 0) :
            return True
        return False

    def markHit(self) :
        self.status = 1
    def markMiss(self) :
        self.status = 2;

    def getStatus(self) :
        return self.status
    def hasShip(self) :
        return self.exist
    def setShip(self, b) :
        self.exist = b
    def setStatus(self, x) :
        self.status = x

class Grid:
    #SIZE = 10
    #grid = np.zeroes((10, 10))

    def __init__(self, size) :
        self.grid = [[None for _ in range(10)] for _ in range(10)]
        self.size = size
        count = -1
        c2 = -1
        for i in range(size) :
            #print(i)
            count+=1
            for j in range(size) :
                self.grid[i][j] = Location()


    def markHit(self, x, y) :
        self.grid[x][y].markHit()
    def markMiss(self, x, y) :
        self.grid[x][y].markMiss()
    def setStatus(self, x, y, z) :
        self.grid[x][y].setStatus(z)
    def getStatus(self, x, y) :
        return self.grid[x][y].getStatus()
    def isAlreadyGuessed(self, x, y) :
        if(self.grid[x][y].isUnguessed) :
            return False
        return True
    def setShip(self, x, y, z) :
        self.grid[1][1].setShip(True)
    def hasShip(self, x, y) :
        return self.grid[x][y].hasShip()
    def get(self, x, y) :
        return self.grid[x][y]

    def addShip(self, S):
        r = S.getRow()
        c = S.getCol()
        l = S.getLength()
        d = S.getDir()
        if(d == 1) :
            for i in range(l) :
                self.grid[r+i][c].setShip(True)
        elif (d == 0) :
            for i in range(l) :
                self.grid[r][c + i].setShip(True)
        self.grid[r][c].setShip(True)

    def printStatus(self) :
        top = "  "
        for i in range(0, 10) :
            #if( i == 1) :
            #    print("  1 ")
            if(i < 9) :
                top += str(i) + " "
            else :
                top += str(i)
        print(top)
        for j in range(self.size) :
            line = "" + chr(j + 65) + " "
            for k in range(self.size) :
                if(self.grid[j][k].getStatus() == 0) :
                    line += "- "
                elif (self.grid.getStatus(j,k) == 1) :
                    line += "X "
                elif (self.grid.getStatus(j,k) == 2) :
                    line += "O "
            print(line)
    def printShip(self) :
        #self.setShip(2,4,True)
        top = "  "
        for i in range(1, 11) :
            #if( i == 1) :
            #    print("  1 ")
            if(i < 10) :
                top += str(i) + " "
            else :
                top += str(i)
        print(top)
        for j in range(self.size) :
            line = "" + chr(j + 65) + " "
            for k in range(self.size) :
                if(self.grid[j][k].hasShip()) :
                    line += "X "
                else:
                    line += "- "
            print(line)

print("Welcome to BattleShip!")
print("Press Enter to Continue..")
wait = input("")
p1 = Grid(10)
ls = [5,4,3,3,2]
cu = 0
while (cu < len(ls)) :

    p1.printShip()
    while True:
        sc = input("Where would you like to place a ship of length " + str(ls[cu]) + "? Would you like the ship to go down or right? \nEnter in a comma separate pair (i.e. \"A,0,down\"): ")
        sc1 = sc.split(",")
        r = int(ord(sc1[0]) - 65)
        c = int(sc1[1])
        sd = sc1[2].lower()
        di = 0
        if("down" in sd and r <= 10 - ls[cu]) :
            di = 1
            break
        if("right" in sd and c <= 10 - ls[cu]) :
            break
        else :
            print("Invalid")
    sh = Ship(ls[cu])
    cu += 1
    sh.setRow(r)
    sh.setCol(c)
    sh.setDir(di)
    p1.addShip(sh)
```
